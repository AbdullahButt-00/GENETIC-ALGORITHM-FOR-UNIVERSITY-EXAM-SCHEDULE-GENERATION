```python
import numpy as np
def check_duplicates(matrix):
    matrix = np.array(matrix)
    if matrix.ndim < 2 or matrix.size == 0:
        return -1  # Return -1 if matrix is not two-dimensional or empty
    
    rows, cols = matrix.shape
    count = 0
    for j in range(cols):
        column_values = matrix[:, j]
        unique_values, counts = np.unique(column_values, return_counts=True)
        if len(unique_values) != rows:
            count += 1
    return count

def hard_constraint_exam_scheduling(time_table,fitness):
    # An exam will be scheduled for each course
    '''
    if len(chromosome.time_table) == len(courses_data):
        chromosome.fitness += 5
    else:
        chromosome.fitness -= 5
    '''
    count=0
    course_data=list(courses_data)
    for i in range(len(course_data)):
        if course_data[i] in time_table:
            count=count+1
        if count==len(course_data):
            fitness += 5
        else:
            chromosome.fitness -= 5
def hard_constraint_student_enrollment(time_table,fitness):
    # A student is enrolled in at least 3 courses.
    # A student cannot give more than 1 exam at a time.
    print(time_table)
    if(check_duplicates(time_table)==0):
        fitness +=5
    else:
        fitness -=5
    '''
    for exam in time-table:

        enrolled_courses = len(student_course_data.get(exam.students_enrolled, []))
        if enrolled_courses >= 3 and len(exam.students_enrolled) <= 1:
            chromosome.fitness += 5
        else:
            chromosome.fitness -= 5
        '''


def hard_constraint_exam_scheduling_time(time_table,fitness):
    # Exam will not be held on weekends
    for exam in time_table:
        if not exam.day.startswith("Saturday") and not exam.day.startswith("Sunday"):
            fitness += 5
        else:
            fitness -= 5

def hard_constraint_exam_time_range(time_table,fitness):
    # Each exam must be held between 9 am and 5 pm
    for exam in time_table:
        if exam.time in ["AM 9:00 - 12:00", "PM 2:00 - 5:00"]:
            fitness += 5
        else:
            fitness -= 5

def hard_constraint_teacher_invigilation(time_table, fitness):
    # Each exam must be invigilated by a teacher.
    # A teacher cannot invigilate two exams at the same time.
    if check_duplicates(time_table) == 0:
        fitness += 5
    else:
        fitness -= 5

    # A teacher cannot invigilate two exams in a row.
    t = np.array(time_table)
    if len(t.shape) < 2:
        return fitness  # Not enough data for the constraint

    r, c = t.shape
    count = 0
    for i in range(r - 1):  # Use range(r-1) to avoid IndexError
        for j in range(c):
            if time_table[i][j] == time_table[i + 1][j]:
                count += 1
    if count == 0:
        fitness += 5
    else:
        fitness -= 5

    return fitness

def check_hard_constraints(time_table,fitness):
    hard_constraint_exam_scheduling(time_table,fitness)
    hard_constraint_student_enrollment(time_table,fitness)
    hard_constraint_exam_scheduling_time(time_table,fitness)
    hard_constraint_exam_time_range(time_table,fitness)
    hard_constraint_teacher_invigilation(time_table,fitness)
```

```python
import random

def selection(population, tournament_size):
    if tournament_size <= 0 or tournament_size > len(population):
        raise ValueError("Roulett wheel  size must be positive and less than or equal to population size")

    selected_parents = []
    for _ in range(len(population)):
        tournament = random.sample(population, tournament_size)
        selected_parent = max(tournament, key=lambda x: x.fitness)
        selected_parents.append(selected_parent)

    return selected_parents
```

```python
# ############
import random
import csv

# Load data from CSV files (courses, student names, teacher names, etc.)
# You've already provided this part in your code

# Define parameters
NUM_STUDENTS = len(student_names)
NUM_COURSES = len(courses_data)
POPULATION_SIZE = 100
MUTATION_RATE = 0.1
NUM_GENERATIONS = 100

# Initialize the population (chromosomes)
def initialize_population():
    population = []
    for _ in range(POPULATION_SIZE):
        chromosome = initialize_chromosome(NUM_COURSES, NUM_STUDENTS)
        population.append(chromosome)
    return population

def initialize_chromosome(num_courses, num_students):
    # Assuming each gene represents a course assignment for a student
    # Initialize a random chromosome (course assignments for all students)
    chromosome = [random.sample(range(num_courses), num_courses) for _ in range(num_students)]
    return chromosome

# Evaluate fitness of a chromosome
def calculate_fitness(self):
        # Implement fitness calculation logic based on constraints and criteria
        check_hard_constraints(self.time_table,self.fitness)
        check_soft_constraints(self)


# Roulette wheel selection
def roulette_wheel_selection(population, fitness_scores):
    total_fitness = sum(fitness_scores)
    probabilities = [score / total_fitness for score in fitness_scores]
    selected_index = random.choices(range(len(population)), weights=probabilities)[0]
    return population[selected_index]

# Single-point crossover
def single_point_crossover(parent1, parent2):
    point = random.randint(1, len(parent1) - 1)
    child1 = parent1[:point] + parent2[point:]
    child2 = parent2[:point] + parent1[point:]
    return child1, child2

# Mutation
def mutate(chromosome, mutation_rate):
    for student_idx in range(len(chromosome)):
        if random.random() < mutation_rate:
            gene1, gene2 = random.sample(range(len(chromosome[student_idx])), 2)
            chromosome[student_idx][gene1], chromosome[student_idx][gene2] = (
                chromosome[student_idx][gene2],
                chromosome[student_idx][gene1],
            )

# Main GA loop
def genetic_algorithm():
    population = initialize_population()

    for generation in range(NUM_GENERATIONS):
        # Evaluate fitness for each chromosome
        fitness_scores = [calculate_fitness(chromosome) for chromosome in population]

        # Select parents using roulette wheel
        parents = [roulette_wheel_selection(population, fitness_scores) for _ in range(POPULATION_SIZE)]

        # Perform crossover
        offspring = []
        for i in range(0, POPULATION_SIZE, 2):
            child1, child2 = single_point_crossover(parents[i], parents[i + 1])
            offspring.extend([child1, child2])

        # Apply mutation
        for chromosome in offspring:
            mutate(chromosome, MUTATION_RATE)

        # Replace old population with new offspring
        population = offspring

    # Get the best solution (highest fitness)
    best_chromosome = max(population, key=calculate_fitness)
    return best_chromosome

# Run the GA
best_solution = genetic_algorithm()
print("Best exam schedule (course assignments for students):")
for student_idx, course_assignments in enumerate(best_solution):
    print(f"Student {student_names[student_idx]}: {', '.join(courses_data[course] for course in course_assignments)}")
```
